```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
A, G = 10, 500
# agents, generations
```

```python
fitnesshistories = np.zeros((A, G))
for i in range(A):
    with open(f'fitnesshistory{i}.npy', 'rb') as f:
        fitnesshistories[i] = np.load(f)
```

```python
plt.figure(figsize=(10,6))
for i in range(A):
    plt.plot(range(G), fitnesshistories[i], label=f"seed_{i}")

plt.title(f"Fitness Curve best creature in population in each of {G} generations")
plt.xlabel("Generations")
plt.ylabel("Fitness")
plt.legend()
plt.savefig('combined.png')
```
